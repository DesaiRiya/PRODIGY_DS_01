```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('.\Documents\population_data_t1\pop_data.csv')
df.head()
```

```python
# Create a bar chart
plt.figure(figsize=(10, 10))
plt.bar(df.gender, df.age)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Distribution of Genders in a Population")
plt.show()
```

```python
# Create a histogram
sn.histplot(data=df, x="age", kde=True, bins=range(min(df["age"].fillna(0).astype(int)), max(df["age"].fillna(0).astype(int)) + 2))
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Ages in a Population")
plt.show()
```
